```python
a = int(input('Enter first number:'))
b = int(input('Enter second number:'))
c = input('Select from: ADD SUB MUL DIV MOD').upper()
if c == 'ADD' or c == '+' :
    print(f'{a}+{b}:', a+b)

elif c == 'SUB':
    print(f'{a}-{b}:', a+b)

elif c == 'MUL':
    print(f'{a}*{b}:', a*b)

elif c == 'DIV':
    print(f'{a}/{b}:', a/b)

elif c == 'MOD':
    print(f'{a}%{b}:', a%b)

else:
    print('Invalid choice!!')
```

```python
x = int(input('Enter value on x axis:'))
y = int(input('Enter value on y axis:'))
if x>0 and y>0:
    print(f'FOR x:{x} y:{y} you are in QUADRANT I')

elif x<0 and y>0:
    print(f'FOR x:{x} y:{y} you are in QUADRANT II')

elif x<0 and y<0:
    print(f'FOR x:{x} y:{y} you are in QUADRANT III')

elif x>0 and y<0:
    print(f'FOR x:{x} y:{y} you are in QUADRANT IV')

elif x==0 and y==0:
    print('You are at the origin.')


```
