```python
a = 0
for i in range(1,51):
    if i%5 == 0:
        a+=1
print('Numbers divisible by 5 (1-50):',a)
```

```python
a = int(input('Enter a number: '))
num = a
sum = 0
while a>0:
    b = a%10
    sum = b + sum
    a = a//10

print(f'sum of {num}: {sum}')
```

```python
a = 123321
num = a
c = 0
while a > 0:
    b = a%10
    c = c*10+b
    a = a//10
if num == c:
    print(f'{num} is a palindrome')
else:
    print(f'{num} is not a palindrome')
    
```
